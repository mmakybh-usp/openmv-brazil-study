```python
# --- Imports ---
import pandas as pd
import matplotlib.pyplot as plt

# --- Leitura do CSV ---
# Se o arquivo estiver no repositório do GitHub:
url = "https://raw.githubusercontent.com/mmakybh-usp/openmv-brazil-study/main/benchmark_basic/results/benchmark_fps.csv"

# Se quiser fazer upload manual no Colab, use:
# from google.colab import files
# uploaded = files.upload()
# df = pd.read_csv("benchmark_fps.csv")

df = pd.read_csv(url, delimiter=';')

# --- Visualizar os dados ---
print(df)

# --- Gráfico ---
plt.figure(figsize=(8,5))
plt.bar(df["operacao"], df["fps_medio"])
plt.xlabel("Operação")
plt.ylabel("FPS médio")
plt.title("Benchmark de FPS no OpenMV H7")
plt.xticks(rotation=20)
plt.show()
```
